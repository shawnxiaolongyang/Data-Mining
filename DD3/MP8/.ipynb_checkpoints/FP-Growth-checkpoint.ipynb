```python
class Node(object):
    def __init__(self, data=-1, count=0, parent=None):
        self._data = data
        self._count = count
        self._parent = parent
        self.children = {}
    
```

```python
class Tree(object):
    def __init__(self,data=-1,parent =None,items =None):
        self._root = Node(data='null', parent=self)
        self._items = items
    def route_add(self, route, Root_of_root, count):
        if len(route) < 1:
            return
        element = route.pop(0)
        if element in Root_of_root.children:
            Node_next = Root_of_root.children[element]
        else:
            Node_new = Node(data = element, parent = Root_of_root) 
            Root_of_root.children.setdefault(element, Node_new)
            Node_next = Node_new
        Node_next.count = Node_next.count + count
        if Node_next not in self._items[element]:
            self._items[element].append(Node_next)
        self.route_add(route = route, Root_of_root = Node_next, count = count)
        return
            
            
```

```python
class Tree_generator(object):
    def __init__(self, routes, min_sup=-1, counts=[], headtable={}):
        self._routes = routes
        self._counts = counts
        self._min_sup = min_sup
        self._items = self.Items_get(self._routes)
        self._items_sorted = self.items_get_sorted(self._items)
        self._items_headtable = self.items_init_headtable(headtable)
        self._tree = self.tree_builder(self._counts)
        
    def Items_get(self, routes):
        items = {}
        for route in routes:
            for element in route:
                items.setdefault(element,0)
                items[element] += 1
        return items
        
    def items_get_sorted(self, items = None):
        Item_sorted = []
        sorted_now = sorted(items.iteritems(), key=lambda X:X[1], reverse = True)
        for element in sorted_now:
            if element[1] >= self._min_sup:
                Item_sorted.append(element[0])
        return Item_sorted    
    
    def routes_get_sorted(self, route):
        Route_sorted = []
        for element in self._items_sorted:
            if element in route:
                Route_sorted.append(element)
        return Route_sorted
        
    def items_init_headtable(self, headtable):
        for item in self._items_sorted:
            headtable.setdefault(item, [])
        return headtable
        
    def tree_builder(self, counts):
        tree = Tree(items = self._items)
        for route in self._routes:
            Route_sorted = self.routes_get_sorted(route)
            if counts:
                count = counts.pop(0)
            else:
                count = 1
            tree.route_add(route = Route_sorted, Root_of_root = tree._root, count = count)
        return tree
    
    
    
```
